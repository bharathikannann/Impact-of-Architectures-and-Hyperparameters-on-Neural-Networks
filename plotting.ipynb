# Plotting accuracies for each model

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
version = 1
path = 'siamese_networks/models/my_model_v' + str(version)

train_accuracy = pd.read_csv(path + '/train_accuracy.csv', header = None)
val_accuracy = pd.read_csv(path + '/val_accuracy.csv', header = None)

epochs_range = [i for i in range(1, len(train_accuracy) + 1)]
plt.figure(figsize=(5,5), dpi = 100)
plt.ylim(0, 90)

plt.plot(epochs_range, train_accuracy, label='Train Set')
plt.plot(epochs_range, val_accuracy, label='Val Set')
plt.legend(loc="best")
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.title('Model Accuracy')

plt.tight_layout()
plt.show()
```
